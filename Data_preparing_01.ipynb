```python
record[0][0] #1 access first array first array element
# take the mean of the marks

record_1=np.mean(record,axis=1)[1] # first method
record_2=np.mean(record[1]) #second method
record_3=record[1].mean() #second method

record_4=np.std(record,axis=1)[1]#take the standard deviation
record_5=np.var(record,axis=1)[1] # take the variance np.sqrt(record_4)
```

```python
data_f=pd.DataFrame({"id":[1,2,3,4,5],
                     "name":[np.nan,'indu','kasun',
                             'nadun','ashen'
                             ],
                     "marks":[45,78,np.nan,52,63]})
data_f
```

```python
data_f.isnull() # return the null values
```

```python
data_f.isnull().sum() # return the how many null values in each column
data_f.marks=data_f.marks.fillna(data_f.marks.mean()) # take the mean of the column 'mark'
# and fill into the nan variable
```
